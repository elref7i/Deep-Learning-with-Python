```python
import pandas as pd 
from sklearn.svm import SVC
data=pd.read_csv('C://Users//Rana//OneDrive - Computer and Information Technology (Menofia University)//Desktop//datasetforqiz//pcadata.csv') #dataframe
X=data.drop('outcome',axis=1)
y=data['outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
classifire=SVC(kernel="linear")
classifire.fit(x_train,y_train)
```

```python
predictions=classifire.predict(x_test)
predictions
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=10)
rf.fit(x_train,y_train)
predict=rf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(y_test,predict)

print(matrix)
#               predicted
#               tn        fp
# Actual        
#               fn       tp
#
```

```python
#accurancy score = tp+tn/(tp+tn+fp+fn)
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,predict)
print("accuracy",acc)
```

```python
#precision = tp/(tp+fp)
from sklearn.metrics import precision_score
pre=precision_score(y_test,predict)
print("precision",pre)
```

```python
#recall = tp/(tp+fn)
from sklearn.metrics import recall_score
rec=recall_score(y_test,predict)
print("recall",rec)
```

```python
#f1-measure=2*(recall*precision)/(recall+precision)
from sklearn.metrics import f1_score
f1=f1_score(y_test,predict)
print("f1-measure",f1)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predict))
```

```python
#ROC
from sklearn.metrics import roc_curve
from matplotlib import pyplot as plt
fpr,tpr,_=roc_curve(y_test,predict)
pyplot.plot(fpr,tpr,linestyle='--',label='orignal')
plt.xlabel('FPR')
plt.ylabel("TPR")
```

```python
from sklearn.metrics import roc_auc_score
print(roc_auc_score(y_test,predict))
```
