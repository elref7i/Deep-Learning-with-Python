```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# create dataset
X, y = make_blobs(
   n_samples=150,centers=3 
)
#print(X)
# plot
plt.scatter(
   X[:, 0], X[:, 1],
   c='white', marker='o',
   edgecolor='black', s=50
)
plt.show()
from sklearn.cluster import KMeans

km = KMeans(
    n_clusters=3
)
y_km = km.fit_predict(X)

y_km
```
