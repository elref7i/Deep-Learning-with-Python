```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\python\machinelearing\geyser.csv") 
df.head()
```

```python
df.plot.kde()
```
